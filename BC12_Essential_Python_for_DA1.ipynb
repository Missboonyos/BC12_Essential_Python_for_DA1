# OOP

```python
class Dog:
  def __init__(self, name, species, age):
    self.name = name
    self.species = species
    self.age = age

  def bark(self):
    print(f"{self.name} is barking 'wang wang'!")

  def birthday(self):
    self.age += 1
```

```python
# attributes
dog1 = Dog("I'Pui", "TH Puntang", 2)
dog2 = Dog("Judy", "Beagle", 1)
```

```python
dog1.bark()
dog2.bark()
```

```python
dog1.name
```

```python
print(f"{dog1.name} is {dog1.species} and he is {dog1.age} year(s) old")
print(f"{dog2.name} is {dog2.species} and she is {dog2.age} year(s) old")
```

```python
print(dog1.age)
dog1.birthday()
print(dog1.age)
```

### HW. Create ATM with at least 5 functions by using OOP

- deposit, withdraw, check balance, OTP, order coffee, pay bill, internet, etc.

```python
class ATM:
  def __init__(self, username, bank, balance):
    self.username = username
    self.bank = bank
    self.balance = balance

  def __str__(self):
    return f"{self.username} uses {self.bank} and has EUR {self.balance} in the account."

  def check_balance(self):
    print(f"Your balance is EUR {self.balance}")

  def deposit(self, amount):
    self.balance += amount
    print(f"Deposit EUR {amount} successfully!")

  def withdraw(self, amount):
    self.balance -= amount
    print(f"Withdraw EUR {amount} sucessfully")

  def order_coffee(self, amount):
    self.balance -= amount
    print(f"Order coffee with EUR {amount} successfully")

  def pay_internet(self, amount):
    self.balance -= amount
    print(f"Pay internet with EUR {amount} successfully")

  def pay_bill(self, amount):
    self.balance -= amount
    print(f"Pay bill with EUR {amount} successfully")

  def donation(self, amount):
    self.balance -= amount
    print(f"You donate EUR {amount} successfully")

  def otp(self, otp):
    print(f"Your OTP is {otp}")
```

```python
lila = ATM("lilapurple", "kbank", 1000)
```

```python
# Superclass
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

## extension
class Employee(Person):
  def __init__(self, name, age, company):
    super().__init__(name, age)
    self.company = company

  def greeting(self):
    print("Hello World!")
```

```python
# context manager
read_data = []

with open("hotel.csv", "r") as file:
  data = csv.reader(file)
  for row in data:
    read_data.append(row)
```

```python
read_data
```

```python
# header
read_data[0]
```

```python
# body data
read_data[1: ]
```

```python
## for loop to calculate avg. price per night
price = []

for row in read_data[1: ]:
  price.append(float(row[4])) ## -1 the last item

avg_price = sum(price) / len(price)
print(f"The average price per night is EUR {avg_price}")
```

```python
## another way to write avg price calculation
price = []

for row in read_data[1: ]:
  price.append(int(row[4])) ## -1 the last item

print(f"Average price per night: {sum(price)/len(price)}")
```

```python
## for loop
import requests
import time

names = []

for i in range(5):
  url = f"https://swapi.info/api/people/{i+1}"
  response = requests.get(url)
  response_json = response.json()
  names.append(response_json["name"])
  print(f"Success: {i+1}")
  time.sleep(1)

print(names)
```

# HW: Get data from any public API

```python
# for loop
import requests
import time

characters = []

for i in range(5):
  url = f"https://swapi.info/api/people/{i+1}"
  response = requests.get(url)
  response_json = response.json()
  data = [response_json["name"],
          response_json["height"],
          response_json["mass"],
          response_json["eye_color"]
          ]
  characters.append(data)
  print(f"Success: {i+1}")
  time.sleep(1)

print(characters)
```

```python
import pandas as pd

starwars = pd.DataFrame(characters, columns=["name", "height", "mass", "eye_color"])
starwars
```
